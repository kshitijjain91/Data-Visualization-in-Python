# Basic Plotting: Introduction to ```matplotlib```



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Plotting two 1-D numpy arrays
x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)

plt.plot(x, y)
plt.show()
````

````python
# can also work with lists, though it converts lists to np arrays behind the scenes
plt.plot([1, 4, 6, 8], [3, 8, 3, 5])
plt.show()
````

````python
# Axis labels and 
plt.plot(x, y)

# x and y labels, and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

# # plt.axis(xmin, xmax, ymin, ymax)
plt.axis([0, 110, 0, 1100])
plt.show()
````

````python
# changing the colors and line type
# color blue, line type circle
plt.plot(x, y, 'b+')

# x and y labels, and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
````

````python
# Plotting multiple lines on the same plot

x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)

plt.plot(x, y, 'r-', x, y**2, 'bs', x, y**3, 'g^')
plt.show()
````

You can see the list of colors and shapes here: https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
